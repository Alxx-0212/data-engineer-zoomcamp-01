Drop the 'ehail_fee' column because it's empty

```python
gt.drop(columns=['ehail_fee'],inplace=True)
```

Convert the 'lpep_pickup_datetime' and 'lpep_dropoff_datetime' columns to datetime format

```python
gt.lpep_pickup_datetime = pd.to_datetime(gt.lpep_pickup_datetime)
gt.lpep_dropoff_datetime = pd.to_datetime(gt.lpep_dropoff_datetime)
```

Generate schemas

```python
print(pd.io.sql.get_schema(gt,name = 'green_tripdata_2019-09',con=engine))
```

```python
print(pd.io.sql.get_schema(tz,name = 'taxi+_zone',con=engine))
```

Create tables 

```python
gt.head(n=0).to_sql(name='green_tripdata_2019-09',con=engine,if_exists='replace')
```

```python
tz.head(n=0).to_sql(name='taxi+_zone',con=engine,if_exists='replace')
```

Define iterator for each dataset

```python
gt_iter = pd.read_csv('./dataset/green_tripdata_2019-09.csv', iterator=True, chunksize=10000)
tz_iter = pd.read_csv('./dataset/taxi+_zone_lookup.csv', iterator=True, chunksize=100000)
```

```python
try:
    while True:
        gt = next(gt_iter)
        gt.lpep_pickup_datetime = pd.to_datetime(gt.lpep_pickup_datetime)
        gt.lpep_dropoff_datetime = pd.to_datetime(gt.lpep_dropoff_datetime)
        gt.drop(columns=['ehail_fee'],inplace=True)

        gt.to_sql(name='green_tripdata_2019-09',con=engine,if_exists='append')
except StopIteration:
    print('No next row.')
```

```python
try:
    while True:
        tz = next(tz_iter)

        tz.to_sql(name='taxi+_zone',con=engine,if_exists='append')
except StopIteration:
    print('No next row.')
```
